```python
# Magic
import matplotlib.pyplot as plt
from samila import GenerativeImage
g = GenerativeImage()
g.generate()
g.plot()
plt.show()
```

```python
# Basic
import math
import random
def f1(x, y):
    result = random.uniform(-1,1) * x**2  - math.sin(y**2) + abs(y-x)
    return result
def f2(x, y):
    result = random.uniform(-1,1) * y**3 - math.cos(x**2) + 2*x
    return result
g = GenerativeImage(f1, f2)
g.generate()
g.plot()
g.seed
plt.show()
```

```python
# Projection
from samila import Projection
g = GenerativeImage(f1, f2)
g.generate()
g.plot(projection=Projection.POLAR)
g.seed
plt.show()
```

```python
# Range
g = GenerativeImage(f1, f2)
g.generate(start=-2*math.pi, step=0.01, stop=0)
g.plot()
g.seed
plt.show()
```

```python
# Color
g = GenerativeImage(f1, f2)
g.generate()
g.plot(color="purple", bgcolor="white", projection=Projection.POLAR)
g.seed
plt.show()
```

```python
# Regeneration
g = GenerativeImage(f1, f2)
g.generate(seed=1018273)
g.plot(projection=Projection.POLAR)
plt.show()
```
